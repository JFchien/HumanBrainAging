```python
import pandas as pd
import numpy as np
import anndata as ad
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import scipy as sc
from statsmodels.stats.multitest import fdrcorrection
import warnings
warnings.filterwarnings("ignore")

plt.style.use('./plot_script/mplstyle_czi')
import sys
sys.path.insert(0, './plot_script/')
import cziplot
dpi=150
```

```python
# meta data
cluster_meta=pd.read_csv('../data/TableS7-cell_metadata.csv.gz',index_col=0).dropna()
cluster_meta['age']=cluster_meta['donor'].str[0]
cluster_meta['sex']=cluster_meta['donor'].str[1]
print(cluster_meta.shape)
gene_meta=pd.read_csv('../data/gencode.v37.annotation.intragenic.bed.gz',
                      names=['chr','start','end','gene_id','strand','tmp','gene_name','fn'],
                      index_col='gene_id', sep='\t')
gene_meta['len']=gene_meta['end']-gene_meta['start']
gene_meta=gene_meta[~gene_meta['chr'].isin(['chrY','chrM'])]
genedict_nametoid=dict(zip(gene_meta.gene_name, gene_meta.index))
genedict_idtoname=dict(zip(gene_meta.index, gene_meta.gene_name))
level2_order=cziplot.level2_order
level2_palette=cziplot.level2_palette
```

```python
# DEs
fcthres=np.log2(1.2)
pthres=0.05

fulldata=pd.read_csv('../data/ageDEall.tsv.gz',sep='\t',index_col=0)
fulldata=fulldata[fulldata['celltype'].isin(cziplot.level2_order)]
fulldata['logFDR']=-np.log10(fulldata['adj.P.Val'])
fulldata=fulldata.join(gene_meta['chr'])

data=fulldata[(fulldata['adj.P.Val']<pthres) & (np.abs(fulldata['logFC'])>fcthres)]

fulldata_df=fulldata[~fulldata['chr'].isin(['chrX','chrY','chrM'])].pivot(columns='celltype',values='logFC').fillna(100)
data_df=data[~(data['chr'].isin(['chrX','chrY','chrM']))].pivot(columns='celltype',values='logFC').fillna(100)
```

## summary barplot

```python
! gdown -O /scratch/ranksum_median_intragenic_mCH_ageDMG.tsv.gz  https://drive.google.com/file/d/1FAcsV636ZdQPEjP_yS1V27n57D5GrYWy/view?usp=drive_link --fuzzy
! gdown -O /scratch/ranksum_median_intragenic_mCG_ageDMG.tsv.gz  https://drive.google.com/file/d/1h3nNKwHEJHdpZ1S8AaE78YRuHLRRyIL_/view?usp=drive_link --fuzzy
! gdown -O /scratch/ranksum_median_intragenic_TSS+-1k_mCH_ageDMG.tsv.gz  https://drive.google.com/file/d/1sxdtfJrUOuf9RZ-vK4mItp_q3r_dlcvv/view?usp=drive_link --fuzzy
! gdown -O /scratch/ranksum_median_intragenic_TSS+-1k_mCG_ageDMG.tsv.gz  https://drive.google.com/file/d/16xt_4PPHUBcecuH5ZR0kRpvlvEcLRBD3/view?usp=drive_link --fuzzy
```

```python
corr_df=pd.DataFrame(columns=['celltype','r','p','region'])

regions=['intragenic_mCH','intragenic_mCG','intragenic_TSS+-1k_mCH','intragenic_TSS+-1k_mCG']

for region in regions:
    mC_data=pd.read_csv(f'/scratch/ranksum_median_{region}_ageDMG.tsv.gz',sep='\t',index_col=0,header=[0,1])
    mC_data=mC_data.xs('log2FC',axis=1,level=1)

    for i,celltype in enumerate(level2_order):
        combine_alldegene=fulldata_df.loc[data_df.index.unique()][[celltype]].join(mC_data[[celltype]],lsuffix='_rna').sort_values(celltype).replace([-np.inf,np.inf,100],np.nan).dropna()
        r,p=sc.stats.spearmanr(combine_alldegene[f'{celltype}_rna'],combine_alldegene[celltype])
        corr_df.loc[len(corr_df)]=[celltype,r,p,region]
    
_,corr_df['fdr']=fdrcorrection(corr_df['p'])
corr_df.sort_values('fdr')
corr_df['sig']=False
corr_df.loc[corr_df['fdr']<0.01,'sig']=True
corr_df=corr_df[corr_df['celltype'].isin(level2_order)]
corr_df['celltype'] = pd.Categorical(corr_df['celltype'], level2_order)
corr_df['region'] = pd.Categorical(corr_df['region'], ['intragenic_mCH','intragenic_mCG','intragenic_TSS+-1k_mCH','intragenic_TSS+-1k_mCG'])
corr_df=corr_df.sort_values(['region','celltype']).reset_index(drop=True)
x_var='region'
y_var='r'
hue_var='celltype'

sig_sorted = corr_df.sort_values([hue_var,'region']).reset_index(drop=True)['sig'].values

plt.figure(figsize=(8,3),dpi=dpi)    
barplot = sns.barplot(x=x_var, y=y_var, hue=hue_var,data=corr_df,palette=level2_palette)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90)

for p,sig in zip(barplot.patches,sig_sorted):
    if sig == True:
        barplot.annotate('*', (p.get_x() + p.get_width()/2.5, 0.01),
                    ha='center', va='center', fontsize=6, color='black')

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
plt.ylim([corr_df[y_var].min()*1.2,corr_df[y_var].max()*1.2])
plt.ylabel('Correlation between\nchange of mC and mRNA\nof the union of age-DE genes (ra)')
plt.grid(axis='y')
barplot.set_axisbelow(True)
plt.title('the union of age-DE genes')
plt.text(-0.45,0.1,'* FDR<0.01',fontsize=6)
```

## scatter plot

```python
region='intragenic_mCH'
mC_data=pd.read_csv(f'/scratch/ranksum_median_{region}_ageDMG.tsv.gz',sep='\t',index_col=0,header=[0,1])
mC_data=mC_data.xs('log2FC',axis=1,level=1)

fig,axs=plt.subplots(1,3,figsize=(4.5,1.8),sharex=True,sharey=True,dpi=150)
for i,celltype in enumerate(['L2-4IT_CUX2','L4-5IT_RORB_TSHZ2','MGE_PVALB']):
    ax=axs[i]
    
    combine_allgene =fulldata_df[[celltype]].join(mC_data[[celltype]],lsuffix='_rna',how='inner').sort_values(celltype).replace([-np.inf,np.inf,100],np.nan).dropna()
    combine_allgene.index = combine_allgene.index.map(genedict_idtoname)
    combine_alldegene=fulldata_df.loc[data_df.index.unique()][[celltype]].join(mC_data[[celltype]],lsuffix='_rna').sort_values(celltype).replace([-np.inf,np.inf,100],np.nan).dropna()
    combine_alldegene.index = combine_alldegene.index.map(genedict_idtoname)
    
    ax.scatter(combine_alldegene[f'{celltype}_rna'],combine_alldegene[f'{celltype}'],s=0.2,color='lightgrey',rasterized=True)
    r_all,p_all=sc.stats.spearmanr(combine_alldegene[f'{celltype}_rna'],combine_alldegene[f'{celltype}'])
    
    if celltype in data_df.columns:
        combine_degene=data_df[[celltype]].join(mC_data[[celltype]],lsuffix='_rna',how='inner').sort_values(celltype).replace([-np.inf,np.inf,100],np.nan).dropna()
        combine_degene.index = combine_degene.index.map(genedict_idtoname)
        if len(combine_degene)>25:
            x=combine_degene[f'{celltype}_rna']
            y=combine_degene[celltype]
            xy = np.vstack([x,y])
            z = sc.stats.gaussian_kde(xy)(xy)
            ax.scatter(x,y,c=z,s=0.3,cmap='viridis',rasterized=True)
            r_de,p_de=sc.stats.spearmanr(combine_degene[f'{celltype}_rna'],combine_degene[f'{celltype}'])
        else:
            ax.scatter(combine_degene[f'{celltype}_rna'],combine_degene[f'{celltype}'],s=0.3,c='#440154FF',rasterized=True)

    ax.set_title(f'{cziplot.celltype_simdic[celltype]}',fontsize=6)
    ax.set_xlim([-5.01,5.01])
    ax.set_xticks([-5,-2.5,0,2.5,5])
    ax.grid()
    r_de=np.nan
    p_de=np.nan
    texts=[]
    combine_degene['tmp']=combine_degene[f'{celltype}_rna']*combine_degene[f'{celltype}']
    genes=combine_degene[combine_degene[f'{celltype}']>0].sort_values(f'tmp').head(3).index.to_list()+combine_degene[combine_degene[f'{celltype}']<0].sort_values(f'tmp').head(4).index.to_list()
    genes=set(genes)
    
    ax.scatter(combine_alldegene.loc[genes][f'{celltype}_rna'],combine_alldegene.loc[genes][f'{celltype}'],s=0.3,color='r')
    if len(genes)>0:
        for txt in genes:
            test=ax.text(combine_alldegene.loc[txt,f'{celltype}_rna'], combine_alldegene.loc[txt,f'{celltype}'],txt, fontsize=6)
            texts.append(test)
    
fig.text(0.5, 0.01, 'Change of mRNA expression log2(Aged / Young)', ha='center', va='center',fontsize=8)
fig.text(0.01, 0.5, 'Change of gene body mCH\nlog2(Aged-Young)', ha='center', va='center', rotation='vertical',fontsize=8)
plt.suptitle(f'DE vs {region}')
plt.tight_layout()
```

# mC in age vs cell type

```python
# ! pip install gdown
# ! gdown -O /scratch/binc_CelltypesDonors_binsize1kb.tsv.gz  https://drive.google.com/file/d/1oHuKOOci0pEjoprB5E6a28vD9F-qVxym/view?usp=sharing --fuzzy
```

```python
%%time
binsize=1000

df_binc=pd.read_csv(f'/scratch/binc_CelltypesDonors_binsize{binsize//1000}kb.tsv.gz',sep='\t',
                    header=[0,1,2,3],index_col=[0,1],
                   )
badbins = df_binc.isna().mean(axis=1)>0.2
dfu = df_binc.loc[~badbins]
dfu = dfu.stack('mCtype')
dfu.columns=dfu.columns.droplevel('donor').reorder_levels(['sample','celltype'])
dfu = dfu.T
young=dfu.index.get_level_values('sample').str.startswith('Y') # Use only the Young samples for celltype differences
dfu_ctype=dfu.loc[young].groupby('celltype').mean()
dfu_cond_ctype = dfu - dfu_ctype
```

```python
def my_range(x):
    return x.max(axis=0)-x.min(axis=0)

dfu_ctype_std = my_range(dfu_ctype).rename('celltype range').to_frame()
dfu_cond_ctype_std = my_range(dfu_cond_ctype).rename('range within celltype').to_frame()
dfu_std = dfu_ctype_std.join(dfu_cond_ctype_std,lsuffix='_ctype',rsuffix='_cond_ctype')
dfu['Age'] = dfu.index.get_level_values('sample').str[0]
dfu_age_diff = dfu.groupby(['Age','celltype']).mean()
dfu_age_diff=(dfu_age_diff.loc['A']-dfu_age_diff.loc['Y']).abs().max(axis=0)
dfu_std['age_diff']=dfu_age_diff
dfu_std=dfu_std.unstack('mCtype')
dfu_std=dfu_std[dfu_std[('celltype range','mCG')]>0.01]
dfu_std=dfu_std.stack('mCtype').reset_index()
```

```python
# Define quantiles to be shown
q_show = np.logspace(-4,-1,20)
q_show = np.sort(np.concatenate((q_show,np.arange(.15,.5,.025),1-q_show)))

for mCtype in ['mCG','mCH']:
  dfu_std.loc[dfu_std['mCtype']==mCtype, 'age_diff_binned_mid']=pd.qcut(dfu_std.loc[dfu_std['mCtype']==mCtype, 'age_diff'],
                                                                        q_show).apply(lambda x: x.mid).astype(float)
dfu_std = dfu_std.set_index(['chr','bin','mCtype'])
dfu_stdu=dfu_std[dfu_std.index.get_level_values('chr')!='chrX']
```

```python
with sns.plotting_context("poster"):
    g=sns.relplot(data=dfu_stdu,
            x='age_diff_binned_mid',y='celltype range',
            col='mCtype',hue='mCtype',
              errorbar=('pi',50),
            kind='line',
            facet_kws={'sharey': False, 'sharex': False},
              height=6,aspect=2,
              legend=False,
           )
    g.set(xlabel='Absolute age difference\nin mC (Aged-Young)')
    g.axes[0,0].set(ylabel='mC range (max-min)\nacross cell types\nin young donors (±IQR)')
    [ax.grid(True) for ax in g.axes.flatten()]
    g.figure.suptitle(f'Bins ({binsize//1000} kb)')
    [ax.set_ylim([0,ax.get_ylim()[1]]) for ax in g.axes.flatten()]
```

```python
with sns.plotting_context("poster"):
    g_hist=sns.displot(data=dfu_stdu,
                       x='age_diff',
                       col='mCtype',col_order=['mCG','mCH'],
                       hue='mCtype',hue_order=['mCG','mCH'],
                height=4,aspect=2,
                  stat='count',
                  kde=True,
                  log_scale=(False,False),
                  legend=False,
                       bins=50,
                  common_bins=False,
                facet_kws={'sharey': False, 'sharex': False},)
    for ax1,ax2 in zip(g_hist.axes.flatten(), g.axes.flatten()):
      ax1.set_xticks(ax2.get_xticks())
      ax1.set_xlim(ax2.get_xlim())
    [ax.grid(True) for ax in g_hist.axes.flatten()]
```

```python
g_hist.set(xlabel='Age difference in mC (Aged-Young)')
g_hist.set(ylabel=f'Number of {binsize//1000}kb bins (density)')
g_hist.set(ylim=[1e4,2.5e5])
g_hist.figure
```
