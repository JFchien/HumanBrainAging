```python
from ALLCools.plot import *
import pandas as pd
import numpy as np
import anndata as ad
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import scipy as sc
from statsmodels.stats.multitest import fdrcorrection
import warnings
warnings.filterwarnings("ignore")

plt.style.use('./plot_script/mplstyle_czi')
import sys
sys.path.insert(0, './plot_script/')
import cziplot
dpi=150
```

```python
# meta data
cluster_meta=pd.read_csv('../data/TableS7-cell_metadata.csv.gz',index_col=0).dropna()
cluster_meta['age']=cluster_meta['donor'].str[0]
cluster_meta['sex']=cluster_meta['donor'].str[1]
print(cluster_meta.shape)
gene_meta=pd.read_csv('../data/gencode.v37.annotation.intragenic.bed.gz',
                      names=['chr','start','end','gene_id','strand','tmp','gene_name','fn'],
                      index_col='gene_id', sep='\t')
gene_meta['len']=gene_meta['end']-gene_meta['start']
gene_meta=gene_meta[~gene_meta['chr'].isin(['chrY','chrM'])]
genedict_nametoid=dict(zip(gene_meta.gene_name, gene_meta.index))
genedict_idtoname=dict(zip(gene_meta.index, gene_meta.gene_name))
level2_order=cziplot.level2_order
level2_palette=cziplot.level2_palette
```

```python
n1=len(cluster_meta[cluster_meta['DNA_passQC']])
n2=len(cluster_meta[cluster_meta['RNA_passQC']])
n3=len(cluster_meta[(cluster_meta['DNA_passQC'])&((cluster_meta['RNA_passQC']))].dropna())
print(f'Total: {len(cluster_meta)} nucleis,\nAsterQC: {n1} DNA and {n2} RNA')
print(f'Overlap between RNA and DNA: {n3}')
```

# cell type markers

```python
genelistall= []
genelistexc=['SATB2',
            'CUX2',
            'RORB','PLCH1','TSHZ2','LRRK1','ARHGAP15',
            'TLE4','ADRA1A','THEMIS','LINC00343','CUX1',
            'FAM95C','NXPH4'
            ]
genelistinh=['GAD1','ADARB2',
             'LAMP5',
             'VIP','ADAM33','PAX6',
            'LHX6','SST','CLMP','NPY',
             'PVALB','COL15A1'
            ]
genelistglia=[
             'GFAP','MOG','CSF1R'
            ]
genelists=list(map(genedict_nametoid.get, genelistall+genelistexc+genelistinh+genelistglia))

fig,axs=plt.subplots(1,2,figsize=(7.5,4),dpi=dpi,sharey=True)

## RNA
RNA_marker_data=RNA_marker_data.merge(cluster_meta[['level2']],left_index=True,right_index=True)

rna_heatmap=RNA_marker_data.groupby('level2').mean()
rna_heatmap=rna_heatmap.loc[cziplot.level2_order_full][genelistall+genelistexc+genelistinh+genelistglia]
ax=axs[0]
sns.heatmap(rna_heatmap,
                cmap="viridis",
                vmin=0, vmax=5,
                cbar_kws = {'orientation': 'horizontal','shrink':0.5},
                ax=ax
                  )
ax.axhline(y=11,c='white',linewidth=0.5)
ax.axhline(y=21,c='white',linewidth=0.5)
ax.axvline(x=14,c='white',linewidth=0.5)
ax.axvline(x=26,c='white',linewidth=0.5)
ax.set(ylabel='Cell types')
ax.set_title('mRNA expression (log2(CPM+1))')


## mCH
mCH_marker_data=mCH_marker_data.merge(cluster_meta[['level2']],left_index=True,right_index=True)

mch_heatmap=mCH_marker_data.groupby('level2').mean()
mch_heatmap=mch_heatmap.loc[cziplot.level2_order_full][genelistall+genelistexc+genelistinh+genelistglia]
ax=axs[1]
sns.heatmap(mch_heatmap,
                cmap="viridis_r",
                vmin=0, vmax=0.06,
                cbar_kws = {'orientation': 'horizontal','shrink':0.5},
                ax=ax
                  )
ax.axhline(y=11,c='white',linewidth=0.5)
ax.axhline(y=21,c='white',linewidth=0.5)
ax.axvline(x=14,c='white',linewidth=0.5)
ax.axvline(x=26,c='white',linewidth=0.5)
ax.set(ylabel=None)
ax.set_title('Gene body DNA methylation (mCH)')
plt.tight_layout()
```

# ADARB2

```python
goi='ADARB2'
df=pd.DataFrame(index=cluster_meta.index)
df['rna']=RNA_marker_data[goi]
df['mc']=mCH_marker_data[goi]

df=df.join(cluster_meta[cluster_meta['level2'].notna()]['level2'],how='inner')

plt.figure(figsize=(2.2,2),dpi=dpi)
sns.scatterplot(data=df,x='rna',y='mc',hue='level2',palette=level2_palette,hue_order=level2_order,s=1,rasterized=True)
plt.legend(bbox_to_anchor=(1.01, 1))
r,p=sc.stats.spearmanr(df.dropna()['rna'],df.dropna()['mc'])
plt.title(f'{goi}\nr = {r:.2f}')
print(r,p)
plt.ylim([0,0.2])
plt.xticks([0,5,10])
plt.yticks([0,0.1,0.2])
plt.ylabel(f'{goi} gene body mCH')
plt.xlabel(f'{goi} mRNA expression (log2CPM)')
```

```python
df=pd.read_csv('/scratch/L2-4IT_CUX2_log2cpm+1.tsv.gz',sep='\t',index_col=0).replace(0,np.nan).dropna()
df=df[cziplot.sample_order_rna]
df=df.loc[set(df.index).intersection(set(gene_meta[(gene_meta['fn'].isin(['protein_coding','lncRNA']))].index))]
df=df.loc[df.std(axis=1).sort_values().tail(10000).index]
d1a='YM2B'
d1b='YM2A'
d2a='YM3A'
vmin=0
vmax=200

fig,axs=plt.subplots(1,4,figsize=(7.5,2),dpi=dpi,sharey=True)
ax=axs[0]
sns.histplot(x=df[d1b],y=df[d1a], bins=100,cbar=True, vmin=vmin, vmax=vmax, cmap='viridis',ax=ax) 
ax.set_xlim([0,15])
ax.set_ylim([0,15])
ax.set_xticks([0,5,10,15])
ax.set_yticks([0,5,10,15])
ax.set_title('Within donor',fontsize=6)
ax.set_xlabel('Young male 2, sample A')
ax.set_ylabel('Young male 2, sample B')
ax.grid()
r,p=sc.stats.pearsonr(df[d1b],df[d1a])
ax.text(0.5, 13.3, f'Pearson r = {r:.2f}',fontsize=6)

ax=axs[1]
sns.histplot(x=df[d2a],y=df[d1a], bins=100,cbar=True, vmin=vmin, vmax=vmax, cmap='viridis',ax=ax) 
ax.set_xlim([0,15])
ax.set_ylim([0,15])
ax.set_xticks([0,5,10,15])
ax.set_yticks([0,5,10,15])
ax.set_title('Between donors with same age/sex',fontsize=6)
ax.set_xlabel('Young male 3, sample A')
ax.grid()
r,p=sc.stats.pearsonr(df[d2a],df[d1a])
ax.text(0.5, 13.3, f'Pearson r = {r:.2f}',fontsize=6)

ax=axs[2]
sns.histplot(x=df['YF1A'],y=df[d1a], bins=100,cbar=True, vmin=vmin, vmax=vmax, cmap='viridis',ax=ax) 
ax.set_xlim([0,15])
ax.set_ylim([0,15])
ax.set_xticks([0,5,10,15])
ax.set_yticks([0,5,10,15])
ax.set_title('Between donors with different sex',fontsize=6)
ax.set_xlabel('Young female 1, sample A')
ax.text(x=df.loc[genedict_nametoid['XIST']]['YF1A'],y=df.loc[genedict_nametoid['XIST']][d1a],s='XIST')
ax.grid()
r,p=sc.stats.pearsonr(df['YF1A'],df[d1a])
ax.text(0.5, 13.3, f'Pearson r = {r:.2f}',fontsize=6)

ax=axs[3]
sns.histplot(x=df['AM1A'],y=df[d1a], bins=100,cbar=True, vmin=vmin, vmax=vmax, cmap='viridis',ax=ax) 
ax.set_xlim([0,15])
ax.set_ylim([0,15])
ax.set_xticks([0,5,10,15])
ax.set_yticks([0,5,10,15])
ax.set_title('Between donors with different age',fontsize=6)
ax.set_xlabel('Aged male 1, sample A')
ax.grid()
r,p=sc.stats.pearsonr(df['AM1A'],df[d1a])
ax.text(0.5, 13.3, f'Pearson r = {r:.2f}',fontsize=6)

plt.suptitle('mRNA expression (log2CPM)',fontsize=8)
plt.tight_layout()
```
