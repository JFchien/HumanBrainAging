```python
from ALLCools.plot import *
import pandas as pd
import numpy as np
import anndata as ad
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import scipy as sc
from statsmodels.stats.multitest import fdrcorrection
import warnings
warnings.filterwarnings("ignore")

plt.style.use('./plot_script/mplstyle_czi')
import sys
sys.path.insert(0, './plot_script/')
import cziplot
dpi=150
```

```python
# meta data
cluster_meta=pd.read_csv('../data/TableS7-cell_metadata.csv.gz',index_col=0)
cluster_meta['age']=cluster_meta['donor'].str[0]
cluster_meta['sex']=cluster_meta['donor'].str[1]
print(cluster_meta.shape)
gene_meta=pd.read_csv('../data/gencode.v37.annotation.intragenic.bed.gz',
                      names=['chr','start','end','gene_id','strand','tmp','gene_name','fn'],
                      index_col='gene_id', sep='\t')
gene_meta['len']=gene_meta['end']-gene_meta['start']
gene_meta=gene_meta[~gene_meta['chr'].isin(['chrY','chrM'])]
genedict_nametoid=dict(zip(gene_meta.gene_name, gene_meta.index))
genedict_idtoname=dict(zip(gene_meta.index, gene_meta.gene_name))
level2_order=cziplot.level2_order
level2_palette=cziplot.level2_palette
```

# QC

```python
fig, axs = plt.subplots(1,6,figsize=(8, 4),dpi=dpi,sharey=True)
plt.locator_params(axis='y', nbins=4)

terms=['FinalDNAReads','DNAOverallMappingRate(%)','mCGFrac','mCHFrac','mCCCFrac']
for i,t in enumerate(terms):
    ax=axs[i]
    sns.violinplot(y='sample',x=t,data=cluster_meta,ax=ax,
                   order=cziplot.sample_order,palette=cziplot.sample_palette,
                   orient="h",linewidth=0.5,width=0.5,scale='width',inner='quartile')
    ax.set(ylabel=' ')

axs[0].set_xlim([250000,3300000])
axs[0].set_xscale('symlog')
axs[0].set_xticks([300000,1000000])
axs[0].set_xlabel('Num. of DNA reads')
axs[1].set_xlim([30,90])
axs[1].set_xlabel('Mapping rate (%)')
axs[2].set_xlim([0.49,1])
axs[2].set_xticks([0.5,0.75,1])
axs[2].set_xlabel('mCG')
axs[3].set_xlim([0,0.21])
axs[3].set_xticks([0,0.1,0.2])
axs[3].set_xlabel('mCH')
axs[4].set_xlim([0,0.031])
axs[4].set_xticks([0,0.015,0.03])
axs[4].set_xlabel('mCCC\n(nominal;due to \nBS non-conversion)')
axs[5].set_xticks([0,1000,2000,3000])
axs[5].grid(axis='x')

axs[0].axvline(x=300000,color='k',linewidth=0.75,linestyle='--')
axs[1].axvline(x=50,color='k',linewidth=0.75,linestyle='--')
axs[2].axvline(x=0.5,color='k',linewidth=0.75,linestyle='--')
axs[3].axvline(x=0.2,color='k',linewidth=0.75,linestyle='--')
axs[4].axvline(x=0.03,color='k',linewidth=0.75,linestyle='--')

df=cluster_meta['sample'].value_counts().to_frame('total')
df['DNA_passQC']=cluster_meta[cluster_meta['DNA_passQC']]['sample'].value_counts()
df=df.loc[cziplot.sample_order]
sns.barplot(y=df.index,x=df["total"],color='white',linewidth=.5, edgecolor="k", ax=axs[5])
sns.barplot(y=df.index,x=df["DNA_passQC"],palette=cziplot.sample_palette,order=cziplot.sample_order,linewidth=.5, edgecolor="k", ax=axs[5])
axs[5].set_xlabel('Num. of nuclei')
```

# cell type composition

```python
df=cluster_meta[cluster_meta['level2'].isin(cziplot.level2_order_full)].groupby(['donor','level2']).size().reset_index()
df.columns=['donor','level2','number']
df=df.pivot(index="level2", columns="donor", values="number")
df=df.divide(df.sum(axis=1),axis=0)
df=df.loc[cziplot.level2_order_full,cziplot.donor_order]
df.reset_index().plot(x='level2', kind='barh', stacked=True,width=.8,
        color=list(map(cziplot.donor_palette.get, cziplot.donor_order)),figsize=(4,6))
plt.legend(bbox_to_anchor=(1.01, 1))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('ratio')
plt.gca().invert_yaxis()
```
